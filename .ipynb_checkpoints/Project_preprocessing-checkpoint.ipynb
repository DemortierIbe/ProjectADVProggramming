```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read CSV files
df_2015 = pd.read_csv('2015.csv')
df_2016 = pd.read_csv('2016.csv')
df_2017 = pd.read_csv('2017.csv')
df_2018 = pd.read_csv('2018.csv')
df_2019 = pd.read_csv('2019.csv')


df_2015.head()
```

```python
# onnodige kolommen verwijderen
columns_to_remove_2015 = ['Standard Error', 'Dystopia Residual']
columns_to_remove_2016 = ['Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual']
columns_to_remove_2017 = ['Whisker.high', 'Whisker.low', 'Dystopia.Residual']  # Corrected columns for 2017

df_2015 = df_2015.drop(columns=columns_to_remove_2015)

df_2016 = df_2016.drop(columns=columns_to_remove_2016)

df_2017 = df_2017.drop(columns=columns_to_remove_2017)

# kolommen bekijken
print("2015 dataset:", df_2015.columns)
print("2016 dataset:", df_2016.columns)
print("2017 dataset:", df_2017.columns)
```

```python
# rows met null waarden verwijderen
df_2015 = df_2015.dropna()

df_2016 = df_2016.dropna()

df_2017 = df_2017.dropna()

df_2018 = df_2018.dropna()

df_2019 = df_2019.dropna()

print("2015 dataset:", df_2015.shape)
print("2016 dataset:", df_2016.shape)
print("2017 dataset:", df_2017.shape)
print("2018 dataset:", df_2018.shape)
print("2019 dataset:", df_2019.shape)
```

```python
#kijken voor duplicates
df_2015 = df_2015.drop_duplicates()

df_2016 = df_2016.drop_duplicates()

df_2017 = df_2017.drop_duplicates()

df_2018 = df_2018.drop_duplicates()

df_2019 = df_2019.drop_duplicates()

# shape controlleren
print("2015 dataset:", df_2015.shape)
print("2016 dataset:", df_2016.shape)
print("2017 dataset:", df_2017.shape)
print("2018 dataset:", df_2018.shape)
print("2019 dataset:", df_2019.shape)
```

```python
# kolommen hernoemen
df_2018 = df_2018.rename(columns={'Country or region': 'Country'})
df_2019 = df_2019.rename(columns={'Country or region': 'Country'})

# kolommen controleren
print("2018 dataset:", df_2018.columns)
print("2019 dataset:", df_2019.columns)
```

```python
# alleen de landen houden die in elke dataset zitten
common_countries = set(df_2015['Country']).intersection(df_2016['Country'], df_2017['Country'], df_2018['Country'], df_2019['Country'])

df_2015 = df_2015[df_2015['Country'].isin(common_countries)]
df_2016 = df_2016[df_2016['Country'].isin(common_countries)]
df_2017 = df_2017[df_2017['Country'].isin(common_countries)]
df_2018 = df_2018[df_2018['Country'].isin(common_countries)]
df_2019 = df_2019[df_2019['Country'].isin(common_countries)]

print("2015 dataset:", df_2015.shape)
print("2016 dataset:", df_2016.shape)
print("2017 dataset:", df_2017.shape)
print("2018 dataset:", df_2018.shape)
print("2019 dataset:", df_2019.shape)
```

```python
import numpy as np

# regio invullen aan de hand van de regio's van de andere datasets
def fill_region_nan(df, df_2015, df_2016):
    country_to_region_2015 = df_2015.set_index('Country')['Region'].to_dict()
    country_to_region_2016 = df_2016.set_index('Country')['Region'].to_dict()
    
    df['Region'] = df['Country'].map(lambda x: country_to_region_2015.get(x, country_to_region_2016.get(x, np.nan)))
    return df

df_2017 = fill_region_nan(df_2017, df_2015, df_2016)
df_2018 = fill_region_nan(df_2018, df_2015, df_2016)
df_2019 = fill_region_nan(df_2019, df_2015, df_2016)

print("2017 dataset:")
print(df_2017.head())
print("\n2018 dataset:")
print(df_2018.head())
print("\n2019 dataset:")
print(df_2019.head())
```

```python
# year kolom toevoegen aan elke dataset
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019

print("2015 dataset:")
print(df_2015.head())
print("\n2016 dataset:")
print(df_2016.head())
print("\n2017 dataset:")
print(df_2017.head())
print("\n2018 dataset:")
print(df_2018.head())
print("\n2019 dataset:")
print(df_2019.head())
```

```python
# alle kolommen hernoemen
df_2017 = df_2017.rename(columns={'Happiness.Score': 'Happiness Score','Happiness.Rank': 'Happiness Rank',
                                  'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
                                  'Health..Life.Expectancy.': 'Health (Life Expectancy)',
                                  'Trust..Government.Corruption.': 'Trust (Government Corruption)'})

df_2018 = df_2018.rename(columns={'Overall rank': 'Happiness Rank', 'Score': 'Happiness Score',
                                  'GDP per capita': 'Economy (GDP per Capita)', 'Social support': 'Family',
                                  'Healthy life expectancy': 'Health (Life Expectancy)',
                                  'Freedom to make life choices': 'Freedom',
                                  'Perceptions of corruption': 'Trust (Government Corruption)'})
df_2019 = df_2019.rename(columns={'Overall rank': 'Happiness Rank', 'Score': 'Happiness Score',
                                  'GDP per capita': 'Economy (GDP per Capita)', 'Social support': 'Family',
                                  'Healthy life expectancy': 'Health (Life Expectancy)',
                                  'Freedom to make life choices': 'Freedom',
                                  'Perceptions of corruption': 'Trust (Government Corruption)'})

```

```python
# kolommen in de zelfde volgorde zetten
column_order = ['Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
                'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
                'Generosity', 'Year']

df_2015 = df_2015[column_order]
df_2016 = df_2016[column_order]
df_2017 = df_2017[column_order]
df_2018 = df_2018[column_order]
df_2019 = df_2019[column_order]

print("Column order for each dataset:")
print("2015 dataset:", df_2015.columns)
print("2016 dataset:", df_2016.columns)
print("2017 dataset:", df_2017.columns)
print("2018 dataset:", df_2018.columns)
print("2019 dataset:", df_2019.columns)

```

```python
# datasets samenvoegen in 1
merged_df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)


print(merged_df.shape)
print(merged_df.head())
```

```python
# afronden tot 3 getallen na de komma
merged_df = merged_df.round(3)

print(merged_df.head())
```

```python
merged_df.head(500)
```

```python
# opslaan als csv
merged_df.to_csv('merged_dataset.csv', index=False)
```

```python
sns.set_style("whitegrid")

# line plot maken
plt.figure(figsize=(10, 6))
sns.lineplot(data=merged_df, x='Year', y='Happiness Score', ci=None)

plt.title('Evolution of Happiness Score (2015-2019)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Happiness Score', fontsize=14)

plt.show()
```

```python
# boxplot maken
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df, x='Year', y='Happiness Score')

plt.title('Distribution of Happiness Score by Year (2015-2019)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Happiness Score', fontsize=14)

plt.show()
```

```python
# grouped barplot voor de verdeling van gelukscore per jaar
plt.figure(figsize=(12, 8))

num_bins = 11  
bin_width = 0.15  

colors = sns.color_palette('muted', n_colors=len(merged_df['Year'].unique()))

positions = np.arange(num_bins)

for i, year in enumerate(merged_df['Year'].unique()):
    year_data = merged_df[merged_df['Year'] == year]
    
    counts, bins = np.histogram(year_data['Happiness Score'], bins=num_bins, range=(0, 10))
    
    plt.bar(positions + (bin_width * i), counts, width=bin_width, align='center', label=year, color=colors[i], alpha=0.7)

plt.title('Distribution of Happiness Score by Year (2015-2019)', fontsize=16)
plt.xlabel('Happiness Score', fontsize=14)
plt.ylabel('Number of Countries', fontsize=14)

plt.xticks(positions + (bin_width * (len(merged_df['Year'].unique()) - 1) / 2), range(0, num_bins))

plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# boxplot per regio per jaar
plt.figure(figsize=(12, 8))

box_plot = sns.boxplot(data=merged_df, x='Year', y='Happiness Score', hue='Region')

plt.title('Boxplot of Happiness Score by Region and Year (2015-2019)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Happiness Score', fontsize=14)

box_plot.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# boxplot van de BBP per regio
plt.figure(figsize=(12, 8))

sns.boxplot(data=merged_df, x='Region', y='Economy (GDP per Capita)')

plt.title('Importance of GDP per Capita for Happiness Score by Region (2015-2019)', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Economy (GDP per Capita)', fontsize=14)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def search_countries_by_happiness_score(df, min_score, max_score):
    
    average_scores_per_country = df.groupby('Country')['Happiness Score'].mean()
    
    filtered_countries = average_scores_per_country[(average_scores_per_country >= min_score) & (average_scores_per_country <= max_score)]
    
    return filtered_countries.index.tolist()

min_score = 5.0
max_score = 6.0
countries_within_range = search_countries_by_happiness_score(merged_df, min_score, max_score)
print("Countries within the happiness score range:", countries_within_range)
```

```python
def search_country_by_name(df, country_name):
    
    filtered_df = df[df['Country'] == country_name]
    
    return filtered_df

country_name = 'Belgium'
country_data = search_country_by_name(merged_df, country_name)
print("Data for", country_name, ":", country_data)
```

```python
import matplotlib.pyplot as plt

def plot_country_happiness_evolution(df, country_name):
    
    country_data = df[df['Country'] == country_name]
    
    plt.figure(figsize=(10, 6))
    plt.plot(country_data['Year'], country_data['Happiness Score'], marker='o', linestyle='-')
    plt.title(f'Evolution of Happiness Score over the Years for {country_name}')
    plt.xlabel('Year')
    plt.ylabel('Happiness Score')
    plt.grid(True)
    plt.show()

country_name = "Belgium"  # Example country name
plot_country_happiness_evolution(merged_df, country_name)
```

```python
def search_countries_by_avg_gdp_range(df, min_avg_gdp, max_avg_gdp):
    
    avg_gdp_per_country = df.groupby('Country')['Economy (GDP per Capita)'].mean()
    
    filtered_countries = avg_gdp_per_country[(avg_gdp_per_country >= min_avg_gdp) & (avg_gdp_per_country <= max_avg_gdp)]
    
    countries_within_avg_gdp_range = filtered_countries.index.tolist()
    
    return countries_within_avg_gdp_range

min_avg_gdp = 1.3
max_avg_gdp = 1.4
countries_within_avg_gdp_range = search_countries_by_avg_gdp_range(merged_df, min_avg_gdp, max_avg_gdp)
print("Countries within the average GDP per capita range:")
print(countries_within_avg_gdp_range)
```

```python
def compare_countries_happiness(df, country1, country2):
    
    country1_data = df[df['Country'] == country1]
    country2_data = df[df['Country'] == country2]

    comparison_df = pd.concat([country1_data, country2_data])
    
    return comparison_df

country1_name = "Denmark"
country2_name = "Switzerland"
comparison_data = compare_countries_happiness(merged_df, country1_name, country2_name)
print("Comparison of happiness scores for", country1_name, "and", country2_name)
print(comparison_data[['Country', 'Year', 'Happiness Score']])
```
